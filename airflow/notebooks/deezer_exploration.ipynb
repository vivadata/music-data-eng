```python
import requests
import pandas as pd
from datetime import date
```

```python
item_type='tracks'
url = f"https://api.deezer.com/chart/0/{item_type}"
resp = requests.get(url)
tracks=resp.json()['data']
tracks
```

```python
today = date.today().isoformat()
rows = []
for t in tracks:
        rows.append({
            "date": today,
            "type": "track",
            "id": t['id'],
            "name": t['title'],
            "artist": t['artist']['name'],
            "rank": t['rank'],
        })
rows        
```

```python
df = pd.DataFrame(rows)
df
```

```python
item_type='artists'
url = f"https://api.deezer.com/chart/0/{item_type}"
resp = requests.get(url)
artists=resp.json()['data']
artists
```

```python
rows = []
for a in artists:
        rows.append({
            "date": today,
            "type": "artist",
            "id": a['id'],
            "name": a['name'],
            "rank": a.get('position', None),  # utiliser la position de l'API
        })
df=pd.DataFrame(rows)
df        
```

```python


from collections import Counter

today = date.today().isoformat()

# Récupérer top 100 tracks
url = "https://api.deezer.com/chart/0/tracks?limit=500"
tracks = requests.get(url).json()['data']

rows = []
for t in tracks:
    artist_id = t['artist']['id']
    # Récupérer l'artiste pour avoir le vrai genre
    artist_resp = requests.get(f"https://api.deezer.com/artist/{artist_id}").json()
    genre = artist_resp.get("genre_id", 0)
    if genre == 0:
        genre_name = "Unknown"
    else:
        genre_name = requests.get(f"https://api.deezer.com/genre/{genre}").json().get("name", "Unknown")
    
    rows.append({
        "track_id": t['id'],
        "track_name": t['title'],
        "artist": t['artist']['name'],
        "rank": t['rank'],
        "genre": genre_name,
    })

df = pd.DataFrame(rows)

# Comptage des genres
df_genres = df['genre'].value_counts().reset_index()
df_genres.columns = ['genre', 'count']
df_genres

```
